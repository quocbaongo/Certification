2. Sử dụng công thức xấp xỉ tích phân tính lại $E[X]$ và $\text{Var}(X)$ ở câu 1.

```python
f = lambda x: x * (12 * x ** 2) * (1 - x)
```

```python
def approximate_integral(f, a, b, eps=.01):
    """
    Args:
        f: function to calculate integral
        [a, b]: domain 
        eps: epsilon
    """
    x = np.arange(a, b, eps)
    y = f(x)
    sum = np.sum(eps * y)
    return sum

a, b = 0, 1
approximate_integral(f, a, b)
```
