12. Viết một hàm số tính đạo hàm của $f(x)$ theo công thức xấp xỉ đạo hàm:

$$f'(x) = \lim_{\epsilon \rightarrow 0} \frac{f(x + \epsilon) - f(x-\epsilon)}{2\epsilon}$$

```python
def derivative(function, value):
  h = 0.00000000001
  top = function(value + h) - function(value)
  bottom = h
  slope = top / bottom
  # Returns the slope to the third decimal
  return float("%.3f" % slope)
```

```python
# For instance
def f(x):
  return x ** 2 + 2 * x + 2

a = derivative(f, 2)
print('Slope of the function at x=%s is %s' % (2, a))
```

13. Sử dụng định lý newton, hãy viết một chương trình xác định nghiệm của hàm $f(x)$ với số vòng lặp tối đa là 100.

```python
def newton(f,Df,x0,epsilon,max_iter):
    '''
    f : function for which we are searching for a root of f(x) = 0
    Df : Derivative of f(x)
    x0 : Initial guess for root of f(x) = 0
    epsilon : stopping if abs(f(x)) < epsilon.
    max_iter : Maximum number of iterations of Newton's method.

    Return: xn: root value
    '''
    xn = x0
    for n in range(0,max_iter):
        fxn = f(xn)
        if abs(fxn) < epsilon:
            print('Found solution after',n,'iterations.')
            return xn
        Dfxn = Df(xn)
        if Dfxn == 0:
            print('Zero derivative. No solution found.')
            return None
        xn = xn - fxn/Dfxn
    print('Exceeded maximum iterations. No solution found.')
    return None
```

15. Viết chương trình tính toán hàm loss function Cross entropy cho đầu vào là hai véc tơ phân phối xác suất dự báo và phân phối xác suất của nhãn ground truth. 

```python
import math
import numpy as np

def func(x):
    hx = 1 / (1 + math.e ** x)
    return hx

print(func(10))
```

```python
def cross_entropy(p, q):

  cross_entropy_sum = 0.0

  for i in range(len(p)):
    element1 = p[i]*math.log(func(q[i]))
    element2 = (1-p[i])*math.log(1 - func(q[i]))

    sum_of_ith = -1 * (element1 + element2)
    cross_entropy_sum += sum_of_ith
  
  return cross_entropy_sum
```

```python
# For instance
p = [0.10, 0.40, 0.50]
q = [0.80, 0.15, 0.05]

print(cross_entropy(p, q))
```
